```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 과거 데이터 준비
```

```python
# 모델 구조 만들기
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
# loss: 분산, mse: 민스퀘어 로스, 표준편차차
```

```python
# 데이터로 학습시키기.
epochs=10000
history = model.fit([1,2,3,4],[5,6,7,8],epochs=epochs)
```

```python
epochs_range = range(epochs)
plt.figure(figsize=(8, 8))
plt.plot(epochs_range, history.history['loss'], label='history')
```

```python
print("예측하기!! ", model.predict([*range(10)]))
epochs=1000
```
